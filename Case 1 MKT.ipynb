```python
masci=masc.groupby('Idade ').agg(Preço = pd.NamedAgg(column='Preço',aggfunc=sum),
                                  Desconto = pd.NamedAgg(column='Desconto/ Cupom ',aggfunc=sum),
                                  Final = pd.NamedAgg(column='Preço final ',aggfunc=sum),
                                  Vendas = pd.NamedAgg(column='Produtos ',aggfunc=len))
masci['Porcentagem de Desconto']=(masci['Desconto']/masci['Preço'])*100
masci
```

```python
femi=fem.groupby('Idade ').agg(Preço = pd.NamedAgg(column='Preço',aggfunc=sum),
                                  Desconto = pd.NamedAgg(column='Desconto/ Cupom ',aggfunc=sum),
                                  Final = pd.NamedAgg(column='Preço final ',aggfunc=sum),
                                  Vendas = pd.NamedAgg(column='Produtos ',aggfunc=len))
femi['Porcentagem de Desconto']=(femi['Desconto']/femi['Preço'])*100
femi
```

```python
zero = df[df['Desconto/ Cupom ']>0]
cont = zero.groupby('Produtos ').agg(Quantidade = pd.NamedAgg(column='Desconto/ Cupom ',aggfunc=len))
cont
```

```python
mascdesc = masc.groupby('Produtos ').agg(Masculino = pd.NamedAgg(column='Gênero ',aggfunc=len))
mascdesc
```

```python
femdesc =fem.groupby('Produtos ').agg(Feminino = pd.NamedAgg(column='Gênero ',aggfunc=len))
femdesc
```

```python
desc=df.groupby('Produtos ').agg(Preço = pd.NamedAgg(column='Preço',aggfunc=sum),
                                  Desconto = pd.NamedAgg(column='Desconto/ Cupom ',aggfunc=sum),
                                  Final = pd.NamedAgg(column='Preço final ',aggfunc=sum),
                                  Vendas = pd.NamedAgg(column='Produtos ',aggfunc=len))
desc['Porcentagem de Desconto']=(desc['Desconto']/desc['Preço'])*100
desc
```

```python
n1 = desc.join(cont['Quantidade'],how='right')
n1
```

```python
n2 = n1.join(mascdesc['Masculino'],how='right')
n2
```

```python
tab1 = n2.join(femdesc['Feminino'],how='right')
tab1
```
